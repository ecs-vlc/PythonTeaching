# Probability

Probability is the measure of the *likelihood* than an event will occur. Probability quantifies as a number in the range of $0$ to $1$, where, $0$ indicates impossibility and $1$ indicates certainty. The higher the probability of an event, the more likely it is that the event will occur. 

Take the example of tossing a fair coin. Since the coin is fair, the two outcomes ('heads' or 'tails') are both equally probable; the probability of heads equals the probability of tails.

$$
P(H)=P(T)=\frac{1}{2}
$$

These concepts are described in the [Kolmogorov axioms](https://en.wikipedia.org/wiki/Probability_axioms), which are a fundamental part of probability theory. In it the probability $P$ of some event $E$, denoted $P(E)$. These assumptions can be summarised as follows: Let $(\Omega, F, P)$ be a measure space with $P(\Omega)=1$. Then $(\Omega, F, P)$ is a probability space, with sample space $\Omega$, event space $F$ and probability measure $P$.

````python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

### First Axiom

The probability of an event is a non-negative real number:

$$
P(E) \in \mathbb{R}, P(E) \ge 0 \qquad \forall E \in F
$$

where $F$ is the event space. In particular, $P(E)$ is always finite.

### Second Axiom

The assumption of unit measure: that the probability that at least one of the elementary events in the entire sample will occur is 1.

$$
P(\Omega)=1
$$

### Third Axiom

The assumption of $\sigma$-additivity.

Any countable sequence of disjoint sets (synonymous with *mutually exclusive* events) $E_1, E_2, \dots, E_m$ satisfies:

$$
P\left(\bigcup_{i=1}^{\infty} E_i \right) = \sum_{i=1}^{\infty} P(E_i)
$$

Applications of probability are wide-ranging; including mathematics, statistics, finance, physics, AI/machine learning, computer science, game theory and more.

## Probabilistic application

Here we consider some of the applicational rules of probability, and the relationships between two potential probabilistic events.

### Independent Events

If two events, $A$ and $B$ are *independent* then the joint probability is:

$$
P(A \, \text{and} \, B)=P(A \cap B)=P(A)P(B)
$$

For example, each coin flip is independent of each other, thus $\frac{1}{2}.\frac{1}{2}=\frac{1}{4}$

### Mutually Exclusive Events

If either event $A$ or event $B$ occurs but never both occurs on a single performance of an experiment, then this is a *mutually exclusive* event.

If two events are mutually exclusive then the probability of *both* occuring is denoted $P(A \cap B)$, or the *intersection*.

$$
P(A \,\text{and}\, B)=P(A \cap B) = 0
$$

If two events are mutually exclusive, then the probability of *either* occuring is denoted as the *union* between the events:

$$
P(A \,\text{or}\, B)=P(A \cup B)=P(A)+P(B)-P(A \cap B) = P(A)+P(B)
$$

where $P(A \cap B)=0$.

### Non-Mutually Exclusive Events

If the events are not mutually exclusive then:

$$
P(A \, \text{or} \, B)=P(A \cup B)=P(A)+P(B)-P(A \cap B)
$$

### Conditional Probability

Conditional probability is the probability of event $A$, given the occurrence of some other event $B$. Conditional probability is denoted $P(A | B)$, and is read 'the probability of A, given B'. This is defined as:

$$
P(A|B)=\frac{P(A \cap B)}{P(B)}
$$

If $P(B)=0$ then $P(A|B)$ is formally undefined by this expression. For example, in a bag of 2 red balls and 2 blue balls, the probability of taking a red ball is $1/2$; however when taking a second ball, the probability of it being either red or blue depends on the ball previously taken.

### Inverse Probability

In probability theory, **Bayes' Rule** relates to the odds of event $A_1$ to event $A_2$, before (prior to) and after (posterior to) *conditioning* on another event $B$. 

$$
P(A|B)=\frac{P(B|A)P(A)}{P(B)}
$$

where $P(A|B)$ is described as the *posterior* distribution, $P(B|A)$ is the probability distribution of some given data, $P(A)$ is the *prior* distribution, and $P(B)$ is the marginal likelihood. In general, the **posterior is proportional to the prior times likelihood**, $P(A|B) \propto P(A)P(B|A)$.

## Introduction via sampling

The simpliest way to introduce probability is via *sampling*. For the most common **coin tossing** scenario, let's say we want to simulate 10 tosses of a coin, with the probability of getting a head being $\frac{7}{10}$:

````python
np.random.binomial(1, .7, 10) #biased coin p = 0.7
````

If you toss a coin 10 times and count the number of heads, you will get:

````python
np.random.binomial(10, .7, 1)
````

If you wish to simulate a coin 10 times, counting the number of heads, and repeating the experiment 5 times, do:

````python
np.random.binomial(10, .7, 5)
````

Here we use the **Binomial distribution**, the probability of obtaining $k$ heads after $N$ tosses of such a coin is:

$$
P(k)=\left(N\atop k\right)p^k q^{N-k},\; k=0, \ldots, N.
$$

where the first two central moments are $\mu=Np$ and $\sigma^2=Npq$.

In probability theory we study the properties of data obtained by a **generative process**. In probabilistic machine learning, we infer properties of the generative process from the **data seen**.

Here we will repeat the same experiment above, tossing a coin 10 times, with 10000 repeats and display the results as a histogram. A histogram is a common way of visualising **distributions** for  the values of a random variable. Commonly, upper case letters $X$ denote variables whereas lower-case letters $x_1, x_2, \dots, x_n$ denote values taken by the variables. $P(X)$ is the distribution of values of the variable, where $P(X=x_1)$ is the probability that the variable $X$ takes the value $x_1$.

````python
x = np.random.binomial(10, .7, 10000)
ls = [i for i in range(11)]
plt.hist(x, color="k", alpha=.3, bins=ls)
plt.show()
````

here we show an **unbiased coin** below to compare to:

````python
x = np.random.binomial(10, .5, 10000)
plt.hist(x, color="k", alpha=.3, bins=ls)
plt.show()
````

A couple of key observations here:

1. Note that within the 10000, we have the mode, median and mean number of samples lying at the value of $np$.
2. The binomial distribution behaves like a discrete-form of the *normal distribution*, except when $p$ is near to $n$ or $0$.

## Central Limit Theorem (CLT)

One of the most useful theorems you will need to know is the [**Central Limit Theorem**](https://en.wikipedia.org/wiki/Central_limit_theorem) (CLT). It's role in statistics and machine learning is to find a systematic way of treating errors. 

We start with a collection of $N$ random numbers $x_k, k=1,\dots,N$ that are values taken by a random variable $X$. Their mean:

$$
\langle X \rangle_N=\frac{1}{N}\sum_{k=1}^N x_k
$$

is also a random variable. If a sample of $N$ random numbers is chosen $S$ times, each of the $S$ samples will have a different mean, which we describe as:

$$
\langle X \rangle_N^{(s)}, s=1,\dots,S
$$

where $(i)$ denotes the $i$th sample. These numbers form a limiting distribution when $N \to \infty$. This tendency of the means of samples is the Central Limit Theorem. If the samples are drawn from a population with a mean of $\mu$ and a standard deviation of $\sigma$, then the mean of the sample means coincides with the population mean, and the standard deviation of the sample means diminishes as the inverse square-root of $N$:

$$
\lim_{N \to \infty}\frac{1}{S}\sum_{s=1}^S \langle x \rangle_N^{(s)}=\mu \\
\lim_{N \to \infty}\frac{1}{S}\sum_{s=1}^S \left( \langle x \rangle_N^{(s)} - \mu \right)^2 = \frac{\sigma^2}{N}
$$

For a sample size $N$, the standard deviation of sample means is $\frac{\sigma}{\sqrt{N}}$.

### Sampling from a uniform distribution

Uniform distributions return samples in the range $a \le x \le b$.

```python
r = np.random.random()
x = (b - a)*r + a
```

The first two moments of the distribution are:

$$
\mu=\frac{a+b}{2}, \qquad \sigma^2=\frac{(a+b)^2}{12}
$$

Here we can see the CLT in action as we increase $N$:

````python
n_samples = [10**2, 10**3, 10**5]
a = 2
b = 4
fig, ax = plt.subplots(ncols=len(n_samples), figsize=(14,4))

for i,N in enumerate(n_samples):
    x = (b-a)*np.random.random((N,)) + a
    ax[i].hist(x, color="k", alpha=.5, bins=10)
    ax[i].set_title(r"$N={}$".format(N))
    ax[i].set_xlabel(r"$\Omega$")
````

If we took many replicates of each of these sample sizes, and took the *mean* over replicates, we would see some convergence to the true population mean. This would in effect represent a Gaussian distribution, formulated mathematically as:

$$
f(x|\mu,\sigma^2)=\frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2 \sigma^2}}=\frac{e^{-\frac{(x-\mu)^2}{2\sigma^2}}}{\sigma\sqrt{2\pi}}
$$

where:

- $\mu$ is the mean of the distribution
- $\sigma$ is the standard deviation

````python
def gaussian_dist(x, slen, p):
    # extract mu - first moment
    mu = p
    # using mu, create sigma.
    sigma = np.sqrt(mu/(12.*slen))
    return np.asarray(np.exp(-(x-mu)**2 / (2*sigma**2)) / (np.sqrt(2*np.pi)*sigma))
````

````python
# some grid points
a = 3.5
b = 4.5
N = 1000
N_reps = 1000
mus = [3.6, 4, 4.4]
sigmas = [5, 10, 20]

# generates a 1-d grid of equally-spaced points, uniform distribution
xv = np.linspace(a, b, N)

fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(14,8))

for j,m in enumerate(mus):
    for i,s in enumerate(sigmas):
        # gaussian-ate
        gauss_xv = gaussian_dist(xv, s**2, m)
        clt_xv = np.random.random((N,N_reps))+(m-((b-a)/2))
        
        ax[0,j].plot(xv, gauss_xv, label=r"$\sigma={}$".format(s))
        ax[1,j].hist(np.mean(clt_xv[:s**2], axis=0), alpha=.3, density=True,
                     label=r"$\sigma={}$".format(s))
        
    ax[0,j].set_xlabel("xv")
    ax[0,j].set_ylabel("gaussian generated")
    ax[1,j].set_ylabel("uniform-mean generated")
    for k in range(2):
        ax[k,j].legend()
        ax[k,j].set_title(r"$\mu={}$".format(m))
plt.tight_layout()
````

## Bayesian Inference

One of the key advantages of a probabilistic approach to modelling data is that it enables you to simulate more data that are distributed like the original data you learned the model from. In this sense, Bayesian models are **generative** models.

### Case Study: Earth's surface water proportion

You throw a small, but accurately drawn, globe in the air and catch it. Every time you catch a blue patch indicating water, you record "W", else, record "L" for land. From a finite sample of globe tosses we want to infer the proportion of the Earth's surface covered with water.

Let the unknown proportion of water be $\theta$. The task is to go from the evidence of a finite sample of the form "WLWWWLWLL..." with the event of "W" occurring considered a success with probability $\theta$ and "L" a failure occurring with probability $1 - \theta$.

Let's write a function where we use **Bayes' Rule** to infer, from an observation, the likely fraction of planetary water coverage. We draw from the binomial probability mass function as an assumption for the distribution of our *data* $P(B|A)$:

$$\genfrac(){0pt}{0}{N}{n_W}\theta^{n_W}(1-\theta)^{N-n_W} $$

This can be generated using `stats.binom.pmf`.

````python
def posterior_grid_approx(grid_points, success, tosses):
    # define a grid; introduce points theta
    # posterior probability, p(theta) is return
    p_grid = np.linspace(0, 1, grid_points)
    # define a prior P(A) - unknown distribution! so use uniform
    prior = np.repeat(1, grid_points) / grid_points
    # compute the likelihood at each point in the grid
    likelihood = stats.binom.pmf(success, tosses, p_grid)
    # compute posterior
    unstd_post = likelihood * prior
    # standardize the posterior, such that P(omega)=1
    posterior = unstd_post / unstd_post.sum()
    # return our grid points, and posterior
    return p_grid, posterior
````

If you are unsure what `stats.binom.pmf` returns, here is a custom functional version:

````python
import math
def binom_custom(x, n, k):
    return math.factorial(n) / (math.factorial(k)*math.factorial(n-k))*(x**k)*(1-x)**(n-k)
````

As you can imagine, some of those factorials could be rather expensive to compute, so I highly recommend you don't call the above function excessively!

````python
point_grid = np.linspace(0, 1, 6)
print(binom_custom(point_grid, 10, 7))
print(stats.binom.pmf(7, 10, point_grid))
````

Given a dataset of 6 W and 3 L, there is a certain probability of generating such a dataset from a binomial distribution with a parameter $\theta$ where $\theta$ is one of the values in `p_grid` in the above example. Using **Bayes' Rule**:

$$
p(\theta|X)=p(\theta|N, n_w)=\frac{p(N, n_w|\theta)p(\theta)}{p(N, n_w)}=\frac{p(N, n_w|\theta)p(\theta)}{\int p(N, n_w|\theta)p(\theta) d\theta}
$$

where $X$ represents our evidence or data. In the above example we use a **uniform prior** $p(\theta)$ is a constant and thus independent of $\theta$. Note that integration over $\theta$ is achieved by scaling by the sum of all grid posteriors, using `sum()`.

````python
n_points = 6
success, tosses = 6, 8
p_grid, post = posterior_grid_approx(n_points, success, tosses)

plt.plot(p_grid, post, 'o-', label="success={}\ntosses={}".format(success,tosses))
plt.xlabel("probability of water")
plt.ylabel("posterior probability")
plt.legend()
print("The maximum posterior probability is at the {}-th entry\n with trial parameter theta={} and p({})={}".format(
    np.argmax(post), p_grid[np.argmax(post)], p_grid[np.argmax(post)],
    np.around(np.max(post),3)
))
plt.show()
````

## Updating probabilities using Bayes' Rule

When you gather more evidence (an addition $N'$ data points with $n_w'$ occurrences of "W") we can use Bayes' Rule with the prior probability of parameter $\theta$ set to the posterior distribution after $N$ data points. The new posterior can be obtained simply:


$$p(\theta| \mbox{new data}, \mbox{old data}) = p(\theta| N', n'_W, N, n_W) = \frac{p(N', n'_W|\theta)p(\theta|N, n_W)}{p(N', n'_W| N, n_W)} = \frac{p(N', n'_W|\theta, N, n_W)p(\theta|N, n_W)}{\int p(N', n'_W|\theta, N, n_W)p(\theta|N, n_W) d\theta} $$

````python
def posterior_grid_from_prior(prior, success, tosses):
    grid_points = len(prior)
    p_grid = np.linspace(0, 1, grid_points)
    # compute likelihood
    likelihood = stats.binom.pmf(success, tosses, p_grid)
    # compute new posterior
    unstd_post = likelihood * prior
    # standardize through integration
    post = unstd_post / unstd_post.sum()
    return p_grid, post
````

If Bayesian updating is to be logically consistent, you would expect that the posterior obtained thus should coincide with the probability inferred by computing the posterior based on looking at all the data at once.  We check this next.

````python
points = 6
w, n = 6, 8
p_grid, posterior = posterior_grid_approx(points, w, n)
plt.plot(p_grid, posterior,'.-', label='success = {}\ntosses = {}'.format(w, n))
wnew, nnew = 2, 4 # 2 additional successes from 4 tosses of the globe
p_grid, posterior = posterior_grid_from_prior(posterior, wnew, nnew)
[w, n] = [w+wnew, n+nnew]
plt.plot(p_grid, posterior, '-', label='success = {}\ntosses = {}'.format(w, n)) # updating total counts
plt.xlabel('probability of water')
plt.ylabel('posterior probability')
plt.title('{} points'.format(points))
plt.legend(loc=0);
````

## Sampling from the posterior

One of the benefits of building probabilistic models is their capacity to *generate* data that resembles the training data from which they were built.  It is in this sense that they are called generative models.  Below we will introduce a set of possible parameter values (the $theta$ that stands for the probability of observing a "W") in *p_grid* (as before) and using all values of $\theta$ in *p_grid* with posterior probabilities (in *posterior*) as computed above we generate samples using the numpy function *np.random.choice*.  (Since the entries in *posterior* store the probability of $\theta$, you may regard these as probabilities of probabilities.)

````python
p_grid, posterior = posterior_grid_approx(grid_points=10, success=7, tosses=12)
nsamples = 100 # The next line returns nsample data points (an observation of a value of $\theta$)
samples = np.random.choice(p_grid, p=posterior, size=int(nsamples), replace=True) 

fig, ax = plt.subplots(1,3, figsize=(12,4))
ax[0].plot(samples, 'o', alpha=0.1)
ax[1].hist(samples,alpha=.5,density=True) 
sns.kdeplot(samples, ax=ax[2])
plt.show()
````

## Multivariate Distributions

For a $p-$dimensional variable $X$, the values it can take are $x_1, \dots, x_p$, where $x_i$ are discrete labels. We can think of each vector in $p$ as a different object we wish to sample from. For a random vector:

$$
\boldsymbol{X} = \left(\begin{array}{c}X_1\\\vdots\\X_p \end{array}\right)
$$

the population mean, or expectation, written as $\mathbb{E}[\boldsymbol{X}]$:

$$\boldsymbol{\mu} = \left(\begin{array}{c}\mu_1\\\vdots\\\mu_p\end{array}\right) = \mathbb{E}[\boldsymbol{X}]= \left(\begin{array}{c}\mathbb{E}[X_1]\\\vdots\\\mathbb{E}[X_p]\end{array}\right), $$

and the variance-covariance matrix is as:

$$
\boldsymbol{\Sigma}=\mathbb{V}[\boldsymbol{X}]=\left(\begin{array}{cccc}
\mathbb{V}[X_1]&\mathsf{Cov}[X_1,X_2]&\cdots&\mathsf{Cov}[X_1,X_p]\\
\mathsf{Cov}[X_2,X_1]&\mathbb{V}[X_2]&\cdots&\mathsf{Cov}[X_2,X_p]\\
\vdots&\vdots&\ddots&\vdots\\
\mathsf{Cov}[X_p,X_1]&\mathsf{Cov}[X_p,X_2]&\cdots&\mathbb{V}[X_p]
\end{array} \right),
$$

Thus generating samples from a **multinomial distribution**; Let $x^{(k)}$ be the outcome of the $k$-th throw of a die, $x^{(k)} \in {1, \dots, p}$. The probability of occureence of the outcomes $(x^{(1)}, \dots, x^{(k)}, \dots, x^{(N)})$ of $N$ rolls of a die can be expressed compactly as:

$$
\boldsymbol{P}_d(\boldsymbol{X}|\boldsymbol{\theta})=\boldsymbol{P}_d(n_1,\dots,n_p|\boldsymbol{\theta})=N! \sum_{j=1}^p \left(\frac{\theta_j^{n_j}}{n_j!} \right)
$$

where drawing samples is written as $\boldsymbol{X} \sim \text{Multinomial}(N, \boldsymbol{\theta})$.

````python
ns=20000
nthrows = [25,  200,  1600, 12800]
probvec = np.asarray([.2,.4,.4])
#sm = []
#x = np.reshape(np.zeros(3*ns),(ns,3))
fig, ax = plt.subplots(nrows=3, ncols=len(nthrows),figsize=(12,12))
for i in range(len(nthrows)):
    sm=np.random.multinomial(nthrows[i],probvec,size=ns)/nthrows[i]
    ax[0][i].scatter(sm[:,0], sm[:,1],marker='.',c='r',alpha=.05/(i+1)**(2))
    ax[0][i].set_aspect(1,'datalim')
    ax[1][i].scatter(sm[:,0], sm[:,2],marker='.',c='r',alpha=.05/(i+1)**(2))
    ax[1][i].set_aspect(1,'datalim')
    ax[2][i].scatter(sm[:,1], sm[:,2],marker='.',c='r',alpha=.05/(i+1)**(2))
    ax[2][i].set_aspect(1,'datalim')
````
