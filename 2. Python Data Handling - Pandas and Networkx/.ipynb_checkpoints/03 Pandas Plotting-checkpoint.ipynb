# 3. Plotting with Pandas

Pandas provides a series of plotting functions which is linked to Matplotlib, but automatically plots data according to the shape of the DataFrame, which is incredibly useful and couples well with *function chaining*.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Normal Plotting

```python
normal = pd.Series(np.random.randn(100))
normal.plot()
```

We can couple this with multiple plots by handing the ax object over to pandas plotting:

```python
fig,ax = plt.subplots(ncols=2, figsize=(12,4))
normal.plot(ax=ax[0])
normal.cumsum().plot(ax=ax[1], grid=True)
```

```python
varibles = pd.DataFrame({'normal': np.random.normal(size=100),
                         'gamma': np.random.gamma(1, size=100),
                         'poisson': np.random.poisson(size=100)})
varibles.cumsum(0).plot()
```

Alternatively we can break up the columns using automatic subplots!

```python
varibles.cumsum(0).plot(subplots=True, grid=True)
```

```python
titanic = pd.read_excel("titanic.xlsx", index_col=0)
titanic.head()
```

```python
titanic.groupby('Pclass').Survived.sum().plot.bar()
```

```python
titanic.groupby(['Pclass',"Sex"]).Survived.sum().plot.barh()
```

```python
death_counts = pd.crosstab([titanic.Pclass, titanic.Sex], titanic.Survived.astype(bool))

fig,ax = plt.subplots(ncols=2, figsize=(12,6))
death_counts.plot.bar(stacked=True, color=['black','gold'], grid=True, ax=ax[0])
(death_counts.div(death_counts.sum(axis=1).astype(float), axis=0)
    .plot.barh(stacked=True, color=['black','gold'],ax=ax[1]))
fig.tight_layout()
```

```python
fig,ax=plt.subplots(ncols=2, figsize=(12,4))
titanic.Fare.plot.hist(bins=20, ax=ax[0], normed=True)
titanic.Fare.dropna().plot.kde(ax=ax[1], style='r--', xlim=(-20,600))
```

```python
titanic.boxplot(column="Fare", by="Pclass")
```

```python
titanic.pivot(values="Age", columns="Survived").plot.kde(xlim=(0,100))
```

```python
wines = pd.read_table("wine.dat", sep='\s+')
wines.columns = [
    "Grape","Alcohol","Malic_Acid","Ash","Alcalinity_of_Ash","Magnesium",
    "Total_Phenols","Flavanoids","Nonflavanoid_phenols","Proanthocyanins",
    "Color_Intensity","Hue","OD280/OD315_of_diluted_wines", "Proline"
]
wines.head()
```

```python
wines.plot.scatter("Color_Intensity","Hue", s=wines.Alcohol*50, alpha=.5)
```

```python
wines.plot.scatter("Color_Intensity","Hue", c=wines.Grape)
```
